# Model Training and Testing

Define the models to train:

```python
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import LinearSVC, SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier

models = {
    "Logistic Regression": LogisticRegression(max_iter= 1000, random_state=42),
    "SVC": SVC(random_state=42, kernel='rbf'),
    "LinearSVC": LinearSVC(random_state=42, max_iter=1000),
    "Decision Tree": DecisionTreeClassifier( random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "KNN": KNeighborsClassifier(weights='distance'),
    "SGD Classifier": SGDClassifier(max_iter=1000, random_state=42),
    "OVR Classifier": OneVsRestClassifier(SVC(random_state=42))

}
```

Evaluate each model using 28-fold cross-validation and store results:

```python
from sklearn.model_selection import StratifiedKFold,cross_val_score

n = 1500
X_flat = x_train[:n].reshape(-1, 32*32)
results = {}
cv = StratifiedKFold(n_splits=28, shuffle=True, random_state=42)
for model_name, model in models.items():
    print("-" * 70)
    print(f"Model: {model_name}")
    cv_scores = cross_val_score(model, X_flat, y_train[:n], cv=cv, scoring='accuracy')
    results[model_name] = {
        "model_name": model_name,
        "scores": cv_scores,
        "mean_accuracy": cv_scores.mean(),
    }
    print(f"{model_name} - Cross-Validation Accuracies: \n{cv_scores.round(4)}")
    print(f"- Max Accuracy: {cv_scores.max():.4f}")
    print(f"- Mean Accuracy: {cv_scores.mean():.4f}")
    print(f"- Min Accuracy: {cv_scores.min():.4f}")
    print("-" * 70)
    
```

See the best model that has the highest mean_accuracy:

```python
current_best_accuracy = 0
for model_name, result in results.items():
    if result["mean_accuracy"] > current_best_accuracy:
        current_best_accuracy = result["mean_accuracy"]
        best_model_name = model_name
        best_model = result["model_name"]

print(best_model_name, "with accuracy of", current_best_accuracy)
```

Let us see if we can get a better results useing the Keras API:

```python
x_train = np.array([i.reshape(32, 32).T for i in TrainImages.to_numpy()]) / 255.0
x_test = np.array([i.reshape(32, 32).T for i in TestImages.to_numpy()]) / 255.0

y_test = TestLabels.to_numpy() - 1
y_train = TrainLabels.to_numpy() - 1

print("x_train.shape =", x_train.shape, "\ny_train.shape =", y_train.shape, "\nx_test.shape =", x_test.shape, "\ny_test.shape =", y_test.shape)
```

Define and build an CNN Sequential model:

```python
import tensorflow as tf
from tensorflow import keras
tf.random.set_seed(42)

model = None
model = keras.Sequential([
keras.layers.InputLayer(shape=(32, 32, 1)),

keras.layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
keras.layers.MaxPooling2D((2, 2)),
keras.layers.BatchNormalization(),

keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
keras.layers.MaxPooling2D((2, 2)),
keras.layers.BatchNormalization(),

keras.layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
keras.layers.MaxPooling2D((2, 2)),
keras.layers.BatchNormalization(),

keras.layers.Conv2D(256, (3, 3), activation='relu', padding='same'),
keras.layers.MaxPooling2D((2, 2)),
keras.layers.BatchNormalization(),

keras.layers.Flatten(),

keras.layers.Dense(256, activation="relu"),
keras.layers.Dropout(0.2),
keras.layers.BatchNormalization(),

keras.layers.Dense(128, activation="relu"),
keras.layers.Dropout(0.2),
keras.layers.BatchNormalization(),

keras.layers.Dense(64, activation="relu"),
keras.layers.Dropout(0.2),
keras.layers.BatchNormalization(),

keras.layers.Dense(32, activation="relu"),
keras.layers.Dropout(0.2),
keras.layers.BatchNormalization(),

keras.layers.Dense(28, activation="softmax")])

model.summary()
```

Compile the model:

```python
model.compile(
	loss="sparse_categorical_crossentropy",optimizer=keras.optimizers.Adam(learning_rate=0.001),metrics=["accuracy"]
)
```

Define and Build a new file path to automatically store the best model after each 'epoch':

```python
checkpoint_filepath = 'best_model.keras'
model_checkpoint_callback = keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_filepath, save_weights_only=False, monitor='val_accuracy', mode='max', save_best_only=True, verbose=1 )
```

Fit the model:

```python
history = model.fit(x_train, y_train, epochs=30, validation_split=0.2, 
                    callbacks=[model_checkpoint_callback])
```

Load and Evaluate the best model:

```python
best_model = keras.models.load_model(checkpoint_filepath)
score = best_model.evaluate(x_test, y_test)
print("Test loss:", score[0])
print("Test accuracy:", score[1]) 
```

Visualize the Accuracy with Epoch:

```python
plt.figure(figsize=(10,5))
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.legend(["accuracy","val_accuracy"])
plt.show()
```

Visualize the Loss with Epoch:

```python
plt.figure(figsize=(10,5))
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.legend(["loss","val_loss"])
plt.show()
```

Use the model to predict:

```python
predictions_Test = best_model.predict(x_test)
predictions_Test = np.argmax(predictions_Test, axis=1)
predictions_Train = best_model.predict(x_train)
predictions_Train = np.argmax(predictions_Train, axis=1)
```

Plot the confusion matrix for both Train and Test sets:

```python
from sklearn.metrics import ConfusionMatrixDisplay

classes = [numeric_arabic[x] for x in range(1, 29)]

plt.figure(figsize=(13, 13))
ConfusionMatrixDisplay.from_predictions(y_train, predictions_Train, display_labels=classes, normalize='true', values_format='.0%', ax=plt.gca())
plt.title("Confusion Matrix for Training Set")
plt.figure(figsize=(13, 13))
ConfusionMatrixDisplay.from_predictions(y_test, predictions_Test, display_labels=classes, normalize='true', values_format='.0%', ax=plt.gca())
plt.title("Confusion Matrix for Test Set")
plt.show()
```
